```python
import duckdb
con = duckdb.connect()
con.register("salaries", df)   # df = the full 93,597-row frame

# Check 1: row count matches the file
con.sql("SELECT COUNT(*) AS n FROM salaries").show()   # expect 93597

# Check 2: reproduce a known median.
# Your verified SE Data Analyst median ($116,700) is the DEDUP'd figure,
# so dedup here or it won't match.
con.sql("""
  SELECT MEDIAN(salary_in_usd) AS med, COUNT(*) AS n
  FROM (SELECT DISTINCT * FROM salaries)
  WHERE experience_level = 'SE' AND job_title = 'Data Analyst'
""").show()   # expect med ≈ 116700
```

```python
con.sql("""
  SELECT job_title, experience_level,COUNT(*) AS n, MEDIAN(salary_in_usd) AS med
  FROM (SELECT DISTINCT * FROM salaries)
  WHERE job_title = 'Data Analyst' OR job_title = 'Data Scientist' OR job_title = 'Data Engineer' OR job_title = 'Machine Learning Engineer' GROUP BY experience_level, job_title ORDER BY med DESC
""").show()   
```

```python
con.sql("""
WITH medians AS (
    SELECT 
        experience_level,
        MEDIAN(CASE WHEN job_title = 'Data Analyst' THEN salary_in_usd END) AS analyst_med,
        MEDIAN(CASE WHEN job_title = 'Data Scientist' THEN salary_in_usd END) AS scientist_med,
        MEDIAN(CASE WHEN job_title = 'Data Engineer' THEN salary_in_usd END) AS engineer_med,
        MEDIAN(CASE WHEN job_title = 'Machine Learning Engineer' THEN salary_in_usd END) AS ml_med
    FROM (SELECT DISTINCT * FROM salaries)
    GROUP BY experience_level
)

SELECT 
    experience_level,
    analyst_med,
    analyst_med - scientist_med AS scientist_median_diff,
    analyst_med - engineer_med AS engineer_median_diff,
    analyst_med - ml_med AS ml_median_diff
FROM medians ORDER BY analyst_med;
""").show()
```

```python
con.sql("""
WITH dupeminus AS (
    SELECT DISTINCT *
    FROM salaries
)
SELECT
    median(CASE WHEN job_title = 'Data Analyst'
                 AND experience_level = 'SE'
                THEN salary_in_usd END) AS analyst_median,

    COUNT(CASE WHEN job_title = 'Data Analyst'
                AND experience_level = 'SE'
               THEN 1 END) AS analyst_count,

    median(CASE WHEN job_title = 'Data Scientist'
                 AND experience_level = 'SE'
                THEN salary_in_usd END) AS scientist_median,

    COUNT(CASE WHEN job_title = 'Data Scientist'
                AND experience_level = 'SE'
               THEN 1 END) AS scientist_count
FROM dupeminus
""").show()
```
